```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

startUpData = pd.read_csv("startup_funding.csv",encoding='utf-8')

startUpData['Year'] = startUpData['Date'].str.slice(start=-4)


freq = startUpData['Year'].value_counts()

#print(freq)

year = freq.index

yearValue = np.array(year.values)
yearValue.sort(axis=0)
investmentCount = []
#print(yearValue)
for i in yearValue:
    if i != '':
        investmentCount.append(freq[i])
        print(i,freq[i])
        
plt.plot(yearValue,investmentCount,marker="o")
plt.show()
```

```python
##Top Indian Cities

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

startUpData = pd.read_csv("startup_funding.csv",encoding='utf-8')

startUpData['CityLocation'].dropna(inplace=True) 

def separateCity(city): 
    return city.split('/')[0].strip()

startUpData['CityLocation'] = startUpData['CityLocation'].apply(separateCity)
startUpData['CityLocation'].replace("Delhi","New Delhi",inplace=True) 
startUpData['CityLocation'].replace("bangalore","Bangalore",inplace=True)

freq = startUpData['CityLocation'].value_counts()

#print(freq)
top_10 = freq[:10]
x = top_10.index

for i in x:
    print(i,top_10[i],sep=" ")
    
plt.pie(top_10,labels=top_10.index,autopct='%.2f%%')

plt.show()

```

```python
##Top Funding Amount

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 


startUpData = pd.read_csv("startup_funding.csv",encoding='utf-8')


startUpData['CityLocation'].dropna(inplace=True) 
startUpData['AmountInUSD'].fillna(0,inplace=True)

def separateCity(city): 
    return city.split('/')[0].strip()

startUpData['CityLocation'] = startUpData['CityLocation'].apply(separateCity)
startUpData['CityLocation'].replace("Delhi","New Delhi",inplace=True) 
startUpData['CityLocation'].replace("bangalore","Bangalore",inplace=True)

freq = startUpData['CityLocation'].value_counts()

def convertToNumeric(amount):
    if amount != 0:
        #print(amount)
        amount = amount.replace(",","")
        #print(amount)
    return int(amount)

startUpData['AmountInUSD'] = startUpData['AmountInUSD'].apply(convertToNumeric)

amountFrq = startUpData.groupby('CityLocation')
label = freq.index

column_names = ['City','Invested-Amount']
resultFrame = pd.DataFrame(columns = column_names)


for i,value in enumerate(label):
    cityValue = {}
    amountValue = amountFrq.get_group(value)
    cityValue['City'] = value
    cityValue['Invested-Amount'] = amountValue['AmountInUSD'].sum()
    resultFrame = resultFrame.append(cityValue,ignore_index=True)

resultFrame = resultFrame.sort_values(by=['Invested-Amount'],ascending=False)

first_10 = resultFrame[:10].values

label = []
amountList = []
for i in first_10:
    label.append(i[0])
    amountList.append(i[1])


perc = np.true_divide(amountList,sum(amountList))*100

for i in range(len(label)):
    print(label[i],format(perc[i],'.2f'),sep=" ")
    
plt.pie(amountList,labels=label,autopct='%.2f%%')

plt.show()
```

```python
##Investment Type

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 


startUpData = pd.read_csv("startup_funding.csv",encoding='utf-8')
startUpData['InvestmentType'].dropna(inplace=True) 
startUpData['AmountInUSD'].fillna(0,inplace=True)

def convertToNumeric(amount):
    if amount != 0:
        #print(amount)
        amount = amount.replace(",","")
        #print(amount)
    return int(amount)

startUpData['AmountInUSD'] = startUpData['AmountInUSD'].apply(convertToNumeric)

startUpData['InvestmentType'].replace("PrivateEquity","Private Equity",inplace=True)
startUpData['InvestmentType'].replace("SeedFunding","Seed Funding",inplace=True)
startUpData['InvestmentType'].replace("Crowd funding","Crowd Funding",inplace=True)


startUpData['InvestmentType'].dropna(inplace=True)
uniqueInvestmentType = startUpData['InvestmentType'].unique()
#print(uniqueInvestmentType)

investmentGrp = startUpData.groupby('InvestmentType')
sumOfInvestment = []
for i in uniqueInvestmentType:
    invesmentValue = investmentGrp.get_group(i)
    sumOfInvestment.append(invesmentValue['AmountInUSD'].sum())
    
#print(sumOfInvestment)

perc = np.true_divide(sumOfInvestment,sum(sumOfInvestment))*100

for i in range(4):
    print(uniqueInvestmentType[i],format(perc[i],'.2f'),sep=" ")

plt.pie(sumOfInvestment,labels=uniqueInvestmentType,autopct='%.2f%%')

plt.show()
```

```python
##Top Industries

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 


startUpData = pd.read_csv("startup_funding.csv",encoding='utf-8')

startUpData['IndustryVertical'].dropna(inplace=True) 
startUpData['AmountInUSD'].fillna(0,inplace=True)
startUpData['IndustryVertical'] = startUpData['IndustryVertical'].str.replace("eCommerce","Ecommerce")
startUpData['IndustryVertical'] = startUpData['IndustryVertical'].str.replace("ECommerce","Ecommerce")
startUpData['IndustryVertical'] = startUpData['IndustryVertical'].str.replace("ecommerce","Ecommerce")

def convertToNumeric(amount):
    if amount != 0:
        #print(amount)
        amount = amount.replace(",","")
        #print(amount)
    return int(amount)

startUpData['AmountInUSD'] = startUpData['AmountInUSD'].apply(convertToNumeric)


industryGroup = startUpData.groupby('IndustryVertical')['AmountInUSD'].sum().sort_values(ascending=False)[:5]

#print(industryGroup)

label = industryGroup.index
amount = industryGroup.values

perc = np.true_divide(amount,sum(amount))*100

for i in range(5):
    print(label[i],format(perc[i],'.2f'),sep=" ")

plt.pie(amount,labels=label,autopct='%.2f%%')
plt.show()


```

```python
##Funding rounds

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 


startUpData = pd.read_csv("startup_funding.csv",encoding='utf-8')

startUpData['StartupName'] = startUpData['StartupName'].str.replace("Flipkart.com","Flipkart")
startUpData['StartupName'] = startUpData['StartupName'].str.replace("Oyo Rooms","Oyo")
startUpData['StartupName'] = startUpData['StartupName'].str.replace("OyoRooms","Oyo")
startUpData['StartupName'] = startUpData['StartupName'].str.replace("Oyorooms","Oyo")
startUpData['StartupName'] = startUpData['StartupName'].str.replace("OYO Rooms","Oyo")
startUpData['StartupName'] = startUpData['StartupName'].str.replace("Paytm Marketplace","Paytm")
startUpData['StartupName'] = startUpData['StartupName'].str.replace("Ola Cabs","Ola")
startUpData['StartupName'] = startUpData['StartupName'].str.replace("Olacabs","Ola")
startUpData['AmountInUSD'].dropna(inplace=True)


def convertToNumeric(amount):
    if amount != 0:
        #print(amount)
        amount = amount.replace(",","")
        #print(amount)
    return int(amount)

startUpData['AmountInUSD'] = startUpData['AmountInUSD'].apply(convertToNumeric)

industryGroup = startUpData['StartupName'].value_counts()

fundingRounds = industryGroup[:5]
companyNameList = fundingRounds.index

for i,value in enumerate(fundingRounds):
    print(companyNameList[i],value,sep=" ")
```
