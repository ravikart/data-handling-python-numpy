```python
import csv

with open('year2017.csv','r') as fileObject:
    fileData  = csv.reader(fileObject,delimiter=',',skipinitialspace=True)
    
    fileList = list(fileData)
    
    wondedPpl = []
    for row in fileList[1:]:
        val = row[10]
        if val != '':
            wondedPpl.append(int(float(val)))
            
            
    #print(wondedPpl)
            
    print(sum(wondedPpl))
```

```python
import csv

with open('year2017.csv','r') as fileObject:
    fileData  = csv.reader(fileObject,delimiter=',',skipinitialspace=True)
    
    fileList = list(fileData)
    
    wondedPpl = []
    for row in fileList[1:]:
        val = row[10]
        countryCol = row[3]
        if val != '' and  countryCol == 'India':
            wondedPpl.append(int(float(val)))
            
            
    #print(wondedPpl)
            
    print(sum(wondedPpl))
```

```python
import csv

with open('year2017.csv','r') as fileObject:
    fileData  = csv.DictReader(fileObject,delimiter=',',skipinitialspace=True)
    
    #print(fileData)
    data = []
    for row in fileData:
        if(row['Weapon_type'] == 'Explosives' and row['casualities'] != ''):
            data.append(int(float(row['casualities'])))
            
    print(sum(data))
    
```

```python
import csv

with open('year2017.csv','r') as fileObject:
    fileData  = csv.DictReader(fileObject,delimiter=',',skipinitialspace=True)
    
    #print(fileData)
    data = {}
    for row in fileData:
        month = row['Month']
        if(row['Month'] != '' and row['Killed'] != ''):
            data[month] = data.get(month,0) + int(float(row['Killed']))
            
    print(data)
    
    for row in data:
        print(row,data[row],sep=" ")
    
```
