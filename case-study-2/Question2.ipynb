```python
##Top Investor

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 


startUpData = pd.read_csv("startup_funding.csv",encoding='utf-8')

startUpData['InvestorsName'].dropna(inplace=True)

investorValue = startUpData.assign(InvestorsName=startUpData.InvestorsName.str.split(',')).explode('InvestorsName')

investorValue['InvestorsName'] = investorValue['InvestorsName'].str.strip()

value = investorValue['InvestorsName'].value_counts()[:5]


for index in range(5):
    print(value.index[index],value.values[index],sep=" ")
    
x_pos = [10,35,60,85,105]
print("Question -2:--")
react = plt.bar(x_pos, value.values,width=10,color="lightblue")
plt.xticks(x_pos, value.index,rotation=90)
plt.ylabel('No. of Investments')
plt.xlabel('Investors')
plt.title('Top Investors')

for i in react:
    height = i.get_height()
    plt.text(i.get_x()+i.get_width()/2.0,.05*height,str(height),ha='center',va='bottom')

plt.show()
```
