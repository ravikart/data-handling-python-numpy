```python
import matplotlib.pyplot as plt
import numpy as np

label = ['a','b','c','d']
sizes = [3,4,6,2]
color = ['blue','purple','yellow','green']
explode = [0.1,0,0,0]
plt.pie(sizes,colors = color,labels=label,autopct="%.2f%%",explode = explode)
plt.axis("equal")
plt.show()

```
