```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,5,0.1)
y = x ** 3
y2 = x ** 2
plt.plot(x,y,linewidth=5,label="x**3")
plt.plot(x,y2,color="red",linewidth=5,label="x**2")
plt.ylabel('x**3')

plt.xlabel('x')
plt.title('Matplot Demo')
#plt.axis([0,10,0,100])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

year = [2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population = [100,120,180,250,300,370]

plt.scatter(year,salary,s=population,c=np.arange(len(year)),edgecolor='b')
plt.show()
```
