```python
## Open and read data file as specified in the question
## Print the required output in given format
import pandas as pd
colnames = ['SepalLength','SepalWidth','PetalLength','PetalWidth ','Species']
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",names=colnames)
data = iris.groupby('Species')

meanValue = data.mean().round(2).values
minValue = data.min().round(2).values
maxValue = data.max().round(2).values
speciesGrp = ['Iris-setosa','Iris-versicolor','Iris-virginica']

for i in range(3):
    #for j in range(3):
    for k in minValue[i]:
        print(format(k,'.2f'),end=" ")
    print(speciesGrp[i])
    for k in maxValue[i]:
        print(format(k,'.2f'),end=" ")
    print(speciesGrp[i])
    for k in meanValue[i]:
        print(format(k,'.2f'),end=" ")
    print(speciesGrp[i])
```
